```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Added to download the missing resource
```

```python
def preprocess_text(text):
    """Lowercase, tokenize, remove stopwords"""
    text = text.lower()
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word.isalpha()]  # keep only words
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return tokens
```

```python
def predict_blood_sugar(age, lifestyle, symptoms):
    # Combine lifestyle + symptoms for NLP
    user_text = lifestyle + " " + symptoms
    tokens = preprocess_text(user_text)

    # Default risk
    risk = "Normal"

    # Define keywords
    low_keywords = ["dizzy", "sweaty", "shaky", "weak", "hungry"]
    high_keywords = ["thirsty", "urinate", "tired", "blurred", "fatigue", "overweight"]

    # Rule-based NLP classification
    if any(word in tokens for word in low_keywords):
        risk = "Low"
    if any(word in tokens for word in high_keywords):
        risk = "High"

    # Age factor (simple rule)
    if age > 40 and risk != "Low":
        risk = "High"

    return risk

# ðŸ”¹ Interactive loop
while True:
    print("\n--- Blood Sugar Checker ---")
    try:
        age = int(input("Enter your age (or 0 to exit): "))
    except ValueError:
        print("Please enter a valid number.")
        continue

    if age == 0:
        print("Exiting Blood Sugar Checker. Stay healthy! ðŸ’™")
        break

    lifestyle = input("Describe your lifestyle (e.g., active, sedentary, exercise, healthy food): ")
    symptoms = input("Describe your symptoms (e.g., thirsty, dizzy, tired, sweaty): ")

    result = predict_blood_sugar(age, lifestyle, symptoms)
    print(f"\nðŸ‘‰ Predicted Blood Sugar Risk Level: {result}")
```
